```python
TOPICS = {
    "sports": ["game", "team", "score", "win", "match", "goal"],
    "finance": ["stock", "market", "investment", "price", "trade"],
    "tech": ["python", "java", "algorithm", "server", "cloud"]
}

def classify(text:str)->str:
  text = text.lower()
  score = {}

  for topic,keyword in TOPICS.items():
    count = 0
    for word in keyword:
      if word in text:
        count += 1;
    score[topic] = count
  best = max(score,key=score.get)
  return best if score[best]>0 else "unknown"

print(classify("Our team won the match with a last-minute goal"))
print(classify("New Python algorithm speeds up data processing"))
print(classify("Hello, how are you?"))
```

```python
sentiment = {
    "Positive":["good", "love", "excellent", "great"],
    "Negative":["bad", "hate", "poor", "worst"]
}

def analyze_review(text):
  text = text.lower()
  score = {}

  for nature,keyword in sentiment.items():
    count = 0
    for word in keyword:
      if word in text:
        count += 1
    score[nature] = count
  bestScore = max(score,key=score.get)
  return bestScore


print(analyze_review("I love this product, great quality!"))
print(analyze_review("Poor quality and bad service."))
```
