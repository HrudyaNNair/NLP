```python
from textblob import TextBlob

text = "Hi, im Happy"
text1 = TextBlob(text)
score = text1.sentiment.polarity
print(score)
```

```python
TOPICS = {
    "sports": ["game", "team", "score", "win", "match", "goal"],
    "finance": ["stock", "market", "investment", "price", "trade"],
    "tech": ["python", "java", "algorithm", "server", "cloud"]
}

def classify(text:str)->str:
  text = text.lower()
  score = {}
  count = 0
  for topic,keyword in TOPICS.items():
    for word in keyword:
      if word in text:
        count += 1;
    score[topic] = count
  best = max(score,key=score.get)
  return best if score[best]>0 else "unknown"

print(classify("Our team won the match with a last-minute goal"))
print(classify("New Python algorithm speeds up data processing"))
print(classify("Hello, how are you?"))
```
