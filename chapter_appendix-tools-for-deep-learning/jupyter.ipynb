```python
def analyze(password):
  score = 0
  suggestions = []
  vulnerubilities = []

  if len(password)<5:
    suggestions.append("Too short")
    vulnerubilities.append("bad")
  elif len(password)<8:
    suggestions.append(" short")
    vulnerubilities.append("bad afghj")
  else:
    score += 2
  if any(c.islower() for c in password):
    score += 1
  else:
    suggestions.append("include lower case character")
  if any(c.isupper()for c in password):
    score += 1
  else:
    suggestions.append("include upper case character")
  if any(c.isdigit()for c in password):
    score += 1
  else:
    suggestions.append("include digits")
  if any(not c.isalnum() for c in password):
    score += 1
  else:
    suggestions.append("include special character")

  common_words = ["password", "admin", "qwerty", "letmein"]
  for word in common_words:
    if word == password.lower():
      suggestions.append("Common word")
      score -= 2

  sequential = ["123", "abc", "qwerty"]
  for word in sequential:
    if word == password.lower():
      suggestions.append("sequential word")
      score -= 2

  if score < 5:
    strength = "weak"
  elif score < 8:
    strength = "Okay"
  else:
    strength = "Strong"
  return {
      "Suggestion":suggestions,
      "Vulnerabilities": vulnerubilities,
      "Strength": strength,
      "Score":score
  }

print(analyze("password123"))
print(analyze("abc123"))
print(analyze("Strong@Pass1"))


```
